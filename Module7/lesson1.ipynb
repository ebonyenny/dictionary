```python
from matplotlib import pyplot as plt
days=[0,1,2,3,4,5,6]
money_spent=[100, 400, 600, 300, 500, 1000, 400]
money_spent_2=[1000, 300, 600, 800, 500, 100, 400]

plt.plot(days, money_spent, color="green", marker='s', linestyle='')
plt.plot(days, money_spent_2, color="red", marker='s', linestyle='')
plt.xlabel('Days of the week')
plt.ylabel('amount spent by my friends and I')
plt.title('My first data visualization graph')
plt.show()

```

```python
from cProfile import label
from turtle import color
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True
columns = ["month_number","facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]
df = pd.read_csv("company_sales_data.csv", usecols=columns)
print("Contents in csv file:\n", df)
plt.plot(df.month_number, df.facecream, color='green', label='facecream')
plt.plot(df.month_number, df.facewash, color='blue', label='facewash')
plt.plot(df.month_number, df.toothpaste, color='red', label='toothpaste')
plt.plot(df.month_number, df.bathingsoap, color='black',
 label='bathingsoap')
plt.plot(df.month_number, df.shampoo, color='yellow', label='shampoo')
plt.plot(df.month_number, df.moisturizer, color='purple', label='moisturizer')
plt.ylabel('amounts')
plt.xlabel('months(in numbers)')
plt.title('Products sold per month')
plt.legend()
plt.show()

```
