```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True
columns = ["month_number","shampoo","moisturizer"]
X_axis = np.arange(len('month_number'))
df = pd.read_csv("company_sales_data.csv", usecols=columns)
plt.bar(X_axis - 0.2, df.shampoo, 0.4, label='Shampoo')
plt.bar(X_axis + 0.2, df.moisturizer, 0.4, label='Mositurizer')
plt.xticks(X_axis, df.month_number)
plt.xlabel("Month")
plt.ylabel("product cost")
plt.title("Sales Comparison between 2 products")
plt.legend()
plt.show()
```
